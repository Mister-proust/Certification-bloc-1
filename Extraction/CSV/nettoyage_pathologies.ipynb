```python
import pandas as pd

df = pd.read_csv("../../data/effectifs.csv", sep=";")
```

```python
df.head()
```

```python
df= df[["annee", "patho_niv1", "patho_niv2", "patho_niv3", "cla_age_5", "sexe", "dept", "Ntop", "Npop" ]]
```

```python
df.head()
```

```python
df_cancer= df[df["patho_niv1"].isin(["Cancers"])]
```

```python
df_cancer.head()
```

```python
df_cancer.groupby("patho_niv2")["Ntop"].sum().sort_values(ascending=False)
```

```python
# Elimination des totaux pour améliorer le calcul puis vérification avec l'année 2020. Nous
# sommes proches de l'attendu donc c'est validé. En effet, les quelques écarts peuvent s'expliquer
# du fait que certaines personnes peuvent avoir plusieurs cancers/plusieurs départements de résidences
# /changer de catégories d'âge

df_2020 = df_cancer[(df_cancer["annee"] == 2020) & (df_cancer["patho_niv2"].notna() ) & (df_cancer["patho_niv3"].notna() ) & (df_cancer["dept"] != "999") & (df_cancer["sexe"] != 9) & (df_cancer["cla_age_5"] != "tsage")]

cancer_counts_2020 = df_2020.groupby("patho_niv2")["Ntop"].sum().sort_values(ascending=False)

print(cancer_counts_2020)
```

```python
#Selon les visualisations, les choix seront fait après. Ce qui est intéressant ici
# c'est uniquement le nombre de patients ayant le cancer ET un cancer affecté d'où la
# suppression des Nan dans ces deux catégories. 
df_final = df_cancer[(df_cancer["patho_niv2"].notna() ) & (df_cancer["patho_niv3"].notna() )]
```

```python
df_final.head()
```

```python
df_final = df_final.reset_index(drop=True)
```

```python
df_final.info()
```

```python
df_final = df_final.rename(columns= {
    "patho_niv1": "Pathologie",
    "annee": "Annee",
    "patho_niv2": "Type_cancer",
    "patho_niv3": "Suivi_patho",
    "cla_age_5": "Classe_age",
    "sexe": "Sexe",
    "dept": "Departement",
    "Ntop": "Effectif_patients",
    "Npop": "Effectif_total"
})

df_final.head()
```
